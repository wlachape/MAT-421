The Euler Method

ODE Initial Value

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def model(x,y,z):
    dydt = -z * x
    return dydt
x0 = 6
y = np.linspace(0,20)
z = 0.5
x1 = odeint(model,x0,y,args=(z,))
z = 1
x2 = odeint(model,x0,y,args=(z,))
z = 1.5
x3 = odeint(model,x0,y,args=(z,))
plt.plot(y,x1,'r-',linewidth=2,label='z=0.5')
plt.plot(y,x2,'b--',linewidth=2,label='z=1')
plt.plot(y,x3,'g:',linewidth=2,label='z=1.5')
plt.xlabel('time')
plt.ylabel('x(y)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline
a = lambda t, s: np.exp(-t) 
b = 0.05
c = np.arange(0, 1 + b, b) 
x1 = -1
x = np.zeros(len(c))
x[0] = x1
for i in range(0, len(c) - 1):
 x[i + 1] = x[i] + b*a(c[i], x[i])
plt.figure(figsize = (12, 8))
plt.plot(c, x, 'bo--', label='Approximate')
plt.plot(c, -np.exp(-c), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('c')
plt.ylabel('a(c)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

```
