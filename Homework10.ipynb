Simpson's Rule

```python
import numpy as np
x = 6
y = np.pi
z = 17
a = (y - x) / (z - 1)
b = np.linspace(x, y, z)
c = np.sin(b)
f1 = (a/6) * (c[0] + 4*sum(c[:z-3:3]) \
 + 4*sum(c[2:z-2:4]) + c[z-2])
f2 = 2 - f1
print(f1)
print(f2)
```

Computing Integrals in Python

```python
import numpy as np
from scipy.integrate import trapz
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import quad

plt.style.use('seaborn-poster')
x = np.arange(0, np.pi, 0.01)
F_exact = -np.cos(x)
F_approx = cumtrapz(np.sin(x), x)
plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} sin(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()

x = 6
y = np.pi
z = 17
a = (y - x) / (z - 1)
b = np.linspace(x, y, z)
c = np.sin(b)
f1 = (a/6) * (c[0] + 4*sum(c[3:z-4]) + c[z-2])
f2 = trapz(c,b)
print(f1)
print(f2)

I_quad, est_err_quad = \
 quad(np.sin, 0, np.pi)
print(I_quad)
err_quad = 2 - I_quad
print(est_err_quad, err_quad)
```
