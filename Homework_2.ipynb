Linear Interpolation 

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

## defining the equation for interpolation
## x values are for person A income
## y values are years

x = [50000, 60000, 55000]
y = [2000, 2010, 2020]
y_out = interp1d(y, x)
x_out = y_out(2007)
print(x_out)
```

Linear Interpolation Graph

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
plt.figure(figsize = (10,8))
plt.plot(y, x, '-ob')
plt.plot(2007, x_out, 'ro')
plt.title('Linear Interpolation for Income in 2007')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```

Cubic Spline Interpolation

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [0, 150000, 100000]
x_eq = CubicSpline(x, y, bc_type='natural')
x_out = np.linspace(0, 2, 100)
y_out = x_eq(x_out)
print(x_out)
```

Cublic Spline Interpolation Graph

```python
plt.figure(figsize = (10,8))
plt.plot(x_out, y_out, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation for Years Since Acquiring a Job and Related Income')
plt.xlabel('Year Since Acquiring a Job')
plt.ylabel('Related Income')
plt.show()
```
